```python
from keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator

import matplotlib.pyplot as plt
import cv2
import random
import numpy as np
```

```python
(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

```python
def plot_images(data):
    random_images = random.choices(data, k=10)
    for image in random_images:
        plt.imshow(image)
        plt.show()
```

```python
def rescale(data):
    rescaled_image = []

    for x in data:
        rescaled = x / 255.0
        rescaled_image.append(rescaled)
    
    rescaled_image = np.array(rescaled_image) 
    print(rescaled_image.shape)
    
    return rescaled_image
```

```python
x_train = rescale(x_train)
x_test = rescale(x_test)
```

```python
plot_images(x_test)
```

```python
x_train = x_train.reshape((-1, 28*28))
x_test =  x_test.reshape((-1, 28*28))
```

```python
print(x_train[0].max(), x_train[0].min())
print(x_test[0].max(), x_test[0].min())
```
